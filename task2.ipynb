```python
# Check for missing values
print(titanic_data.isnull().sum())

# Handle missing values for 'Survived'
titanic_data['Survived'].fillna(titanic_data['Survived'].mode()[0], inplace=True)

# Confirm changes
print(titanic_data.info())
```
